As in both the dataframes country code is common , so we can merge both the dataset

```python
merged_df = pd.merge(df, df_country, how='left', on ="Country Code")
merged_df.head()
```

```python
merged_df.columns
```

```python
#lets findout howmany different countries are there and their respective data
merged_df.Country.value_counts()
```

```python
#lets build a pie chart for this
country_val = merged_df.Country.value_counts().values
country_names = merged_df.Country.value_counts().index
plt.pie(country_val, labels=country_names)
plt.show()
```

#### Format: '%[width].[precision]f%%'
| Part        | Meaning                                         |
| ----------- | ----------------------------------------------- |
| `%` (start) | Start of format specifier                       |
| `1`         | Minimum field width (not very important here)   |
| `.2f`       | Floating point number with **2 decimal places** |
| `%%`        | Escaped `%` symbol (shows as `%` in output)     |


```python
#as the previous pie chart looks very clumsy we'll find the top 5 country using zomato services
plt.pie(country_val[:5], labels=country_names[:5], autopct='%1.2f%%')
plt.show()
```

### Observation
1. The maximum records of zomato is from India and then after United states, then UK.

### Group by
In pandas, groupby(['col1', 'col2', ...]) is used to group a DataFrame based on unique combinations of values across two or more columns.
This allows you to perform aggregate operations (like sum, mean, count, etc.) on each group independently.

```python
ratings = merged_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

### Observation
1. When rating is between 4.5 to 4.9 it indicates ---> excellent
2. When rating is between 3.5 to 4.4 it indicates ---> Very good
3. When rating is between 3 to 3.4 it indicates ---> good
and so on... 

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x = "Aggregate rating", y = "Rating count", data=ratings)
plt.show()
```

```python
sns.barplot(x = "Aggregate rating", y = "Rating count", hue='Rating color', data=ratings, palette=['blue', 'red', 'orange', 'yellow', 'green', 'green'])
# as white won't be visible so i changed it to blue
plt.show()
```

### Observation
1. Not rated count is very high than other ratings
2. Maximum numbers of ratings are between 2.5 to 3.4

```python
# count plot
sns.countplot(x= 'Rating color', data = ratings, hue='Rating color', palette=['blue', 'red', 'orange', 'yellow', 'green', 'green'])
plt.show()
```
